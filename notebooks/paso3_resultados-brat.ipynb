```python
def plotSolutions(p, img):

    z = np.copy(img)
    cs = int(9)

    a = p.reshape(-1, 2)
    for b in a:
        c = b.astype(int)
        z[tuple(c)] = 2
        z[tuple([c[0]+1,c[1]])]=cs
        z[tuple([c[0]+1,c[1]+1])]=cs
        z[tuple([c[0],c[1]+1])]=cs
        z[tuple([c[0]-1,c[1]-1])]=cs
        z[tuple([c[0]-1,c[1]])]=cs
        z[tuple([c[0]-1,c[1]+1])]=cs
        z[tuple([c[0],c[1]-1])]=cs
        z[tuple([c[0]+1,c[1]-1])]=cs
        

    return z

```
